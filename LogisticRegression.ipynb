### Load Basic Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset("iris")
```

```python
data.head()
```

```python
data1=data[data["species"]!="setosa"]
```

```python
data1["species"].unique()
```

```python
data1["species"]=data1["species"].map({"versicolor":0,"virginica":1})
```

```python
data1
```

```python
data1["species"].unique()
```

### Divide in input and output

```python
x=data1.iloc[:,:-1]
y=data1.iloc[:,-1]
```

```python
x.head(2)
```

```python
y.head()
```

### Data Splition

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.head()
```

```python
y_train.head()
```

### Model  Building Part

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model=LogisticRegression()
logistic_model.fit(x_train,y_train)      # 1/1+e^-z==> z=m1x1+m2x2+m3x3+m4x4+C
```

```python
y_pred=logistic_model.predict(x_test)
```

```python
y_test,y_pred
```

```python
x_pred=logistic_model.predict(x_train)
```

### Model Evaluation

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print("Training Accuracy: ",accuracy_score(y_train,x_pred))
print("Testing Accuracy: ",accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
